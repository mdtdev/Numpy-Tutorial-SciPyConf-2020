```python
from numpy.random import rand
```

```python
N = 50
xx = np.linspace(0, 10, N)
e = rand(N)*5.0
y1 = xx + e
```

```python
areas = rand(N)*300
plt.scatter(xx, y1, s=areas)
colors = rand(N)
plt.scatter(xx, y1, s=areas, c=colors)
plt.colorbar()
plt.title("Random scatter")
```
